```python
df = crime.drop(labels = ['modus_operandi', 'weapon_code', 'weapon_description',
                          'crime_code_2', 'crime_code_3', 'crime_code_4', 'cross_street'], axis = 1)
df
```

```python
# Mengubah data types yang dari Object ke dalam bentuk DateTime
#Update datatypes
df['date_reported'] = pd.to_datetime(df['date_reported'], format = '%Y-%m-%d')
df['date_occurred'] = pd.to_datetime(df['date_occurred'])
```

```python
# Membuat column baru yang tidak memiliki time atau jam hanya sampai pada date nya
df['date_occurred_no_time'] = df['date_occurred'].dt.date
```

```python
df
```

```python
# Disini kita menghitung jumlah crime pada setiap harinya dan didalamnya crime apa yang dilakukan
# jadi disini kita groupby date kemunculan setiap tanggal nya dan crime description nya

crime_count_per_day = df.groupby(['date_occurred_no_time', 'crime_description']).size()
crime_count_per_day
```

```python
# Disini kita mengambil data top 20 crime yang terjadi berdasarkan tipe setiap crime nya / crime descriptionnya
# kita juga menampilkan average_daily_count nya dari setiap tipe crime tersebut

top_20_crimes = crime_count_per_day.groupby('crime_description').mean()\
                    .reset_index(name = 'average_daily_count').sort_values(by = 'average_daily_count', ascending = False).head(20)
top_20_crimes
```

```python
# Kita urutkan secara ascending (kecil ke besar) berdasarkan data "average_daily_count" nya

top_20_crimes_sorted = top_20_crimes.sort_values(by = 'average_daily_count', ascending = True)
top_20_crimes_sorted
```

```python
# Kita tampilkan dalam grafik bar plot dari crime description dan jumlah rata" kejadian setiap harinya
plt.barh(top_20_crimes_sorted['crime_description'], top_20_crimes_sorted['average_daily_count'])
```

```python
# Kita disini ingin menampilkan bagaimana jarak dari sebuah kasus itu terjadi dan samapi dia di laporkan
# Show difference in reported and occured - How long it takes to report?
df['Time to report'] = (pd.to_datetime(df['date_reported']) - pd.to_datetime(df['date_occurred_no_time'])).dt.days
```

```python
# Kita menampilkan data yang digabungkan dari setiap crime description dan time to reportnya
# Jadi disini kita menampilkan data jumlah rata" nilai penggabungan setiap crime description dan time to reportnya
# Jadi misalkan data crime description "BOMB SCARE" punya rata" time to reportnya berapa lama

df.groupby(['crime_description'])['Time to report'].mean().sort_values().head(10)
```

```python
df_age.groupby('crime_description')['victim_age'].mean()
```

```python
# Fungsi reset_index pada Pandas digunakan untuk mengatur ulang indeks DataFrame ke urutan default 
# (0, 1, 2, ...). Indeks lama dapat dijadikan kolom baru atau dibuang.

df_age.groupby('crime_description')['victim_age'].mean().reset_index()
```

```python
# Kita menentukan area yang tingkat crime nya tertinggi saat malam hari
# Find the area with the highest number of crimes occurring during nighttime

nighttime_crimes = df[(df['date_occurred'].dt.hour >= 20) | (df['date_occurred'].dt.hour <= 3)]
nighttime_crimes
```

```python
# Kita menjumlahkan jumlah crime yang terjadi pada setiap area saat malam hari

nighttime_crimes['area_name'].value_counts()
```

```python
# Data crime pada setiap jam

hourly_crime = df['date_occurred'].dt.hour.value_counts().sort_index().reset_index()
hourly_crime
```

```python
# Menampilkan dalam bar plot data crime setiap jam nya

plt.bar(hourly_crime['date_occurred'], hourly_crime['count'])
```

```python
# Kita menampilkan sejumlah crime description yang terjadi pada jam 12

df[df['date_occurred'].dt.hour == 12]['crime_description'].value_counts()
```

```python
# df['crime_description'] == 'THEFT OF IDENTITY': Mengecek baris mana yang memiliki deskripsi 'THEFT OF IDENTITY' 
# (menghasilkan nilai boolean True atau False).
# ~: Operator negasi/Tilde, membalikkan hasil. Jadi, True menjadi False dan sebaliknya.
# df[...]: Memilih baris di mana kondisi di dalamnya bernilai True
# Menghasilkan DataFrame baru no_identity_theft yang tidak mengandung data 
# kejahatan dengan deskripsi "THEFT OF IDENTITY".


no_identity_theft = df[~(df['crime_description'] == 'THEFT OF IDENTITY')]
no_identity_theft
```

```python
# no_identity_theft['date_occurred'].dt.hour
# Mengambil jam dari kolom 'date_occurred', yang diasumsikan bertipe datetime.
# .value_counts()
# Menghitung jumlah kejadian untuk setiap jam (misalnya, jam 0, jam 1, dst.).
# .sort_index()
# Mengurutkan hasil berdasarkan jam (indeks) agar berurutan dari jam 0 hingga jam 23.
# .reset_index()
# Mengatur ulang indeks hasil menjadi default (0, 1, 2, ...) dan menyimpan indeks lama sebagai kolom.

new_hourly_count = no_identity_theft['date_occurred'].dt.hour.value_counts().sort_index().reset_index()
```

```python
# Menampilkan dalam bar plot new hourly count dari setiap kemunculan crime 

plt.bar(new_hourly_count['date_occurred'], new_hourly_count['count'])
```
